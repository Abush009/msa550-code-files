## Numbers

```python
#integer operations

a = 5
b = 2

print("a + b = ", a + b)
print("a - b = ", a - b)
print("a * b = ", a * b)
print("a / b = ", a / b)

print("a ** b = ", a ** b) #Exponential
print("a % b = ", a % b) #Modulus



#float operations
x = 3.14
y = 2.71

print("x + y = ", x + y)
print(f"Round x to 1 decimal place: {round(x,1)}" )



#using math module

import math

print(f"Square root of 16: {math.sqrt(16)}")
print(f"Sin of 90 degrees: {math.sin(math.pi/2)}")
```

## Strings
Strings in Python are immutable sequence of characters. Python provides many built-in methods for string manipulation.

```python
# String creation and basic operations

s1 = "Hello"
s2 = "World"
s3 = """
This is a multi-line string.
It can span multiple lines.
"""


print(s1 + " " + s2)
print(s1 * 3)


# String indexing and slicing
text = "Python Programming"
print(f"First character: {text[0]}")
print(f"Last character: {text[-1]}")
print(f"Slicing: {text[7:14]}")
print(f"Reverse: {text[::-1]}")


# Other String methods
print(f"Uppoercase: {text.upper()}")
print(f"Lowercase: {text.lower()}")
print(f"Split: {text.split()}")
print(f"Replace: {text.replace('Programming', 'Coding')}")


# String Formatting

name = "Alice"
age = 25


print("My name is " + name + " and I am " + str(age) + " years old.")
print(f"My name is {name} and I am {age} years old.")
print("My name is {} and I am {} years old.".format(name, age))




```

## Conditional Statements

Conditional statements in python allow you to execute different code blocks based on whether a conditon is true or false. Python uses if, elif, and else for constructing conditional logic.

```python
# Basic if-else statement

age = 21

if age >= 18:
  print("You are an adult.")
else:
  print("You are a minor.")





# if-elif-else statement

score = 85

if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
else:
  print("Grade: D")



# Nested if statements

x = - 10
y = -5

if x > 0:
  if y > 0:
    print("Both x and y are positive.")
  else:
    print("x is positive, but y is not.")
else:
  print("x is not positive.")


is_adult = "Yes" if age >= 18 else "No"
print(f"Is the person an adult? {is_adult}")
```

## Iterations
Iteration in Python allows you to repeat a block of code muultiple times. In Python we can use for loops and while loop to iterate.

```python
# For loop with range

for i in range(5):
  print(i)

#food list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
  print(fruit)

count = 0

while count < 5:
  print(count)
  count += 1

  #Be mindful of infinite loops


# Nested Loops

for i in range(3):
  for j in range(3):
    print(f"({i}, {j})", end=' ')
  print()

#looking at only one index
for i in range(3):
  for j in range(3):
    print(f"{j}", end=' ')
  print()


# Loop control statements
for num in range(10):
  if num ==5:
    continue # skip 5
  if num == 8:
    break #stop iterating beyond this point
  print(num, end=' ')
print()

# List Comprehension
squares = [x**2 for x in range(100)]
print(squares)
```

## Tuples
Tuples are immutable sequences in Python, typically they are used to store collections of heteregenous data. They are defined using parenthesis ().

```python
# creating a tuple

empty_tuple=()
single_item_tuple = (1,)
mixed_tuple = (1, "hello", 2.5)

# Accessing elements
print(f"The second item in the mixed_tuple is: {mixed_tuple[1]}")

#Tuple unpacking
a, b, c = mixed_tuple
print(f"a: {a}, b: {b}, c: {c}")

#Tuples are immutable

#mixed_tuple[0] = 2 # this will raise an error -- tuple object does not support item assignment

numbers = (1, 2, 3, 4, 5, 4, 2, 1)
print(f"Count of 2 in numbers: {numbers.count(2)}")
print(f"Index of 4 in numbers: {numbers.index(4)}")


locations = {
     (35.534534, 46.56756) : "New York City",
     (55.564, 1.435643) : "London"}


print(f"City at (35.534534, 46.56756): {locations[(35.534534, 46.56756)]}")
```
