## Variables
Variables in Python are dynamically typed, meaning you do not have to explicitely declare their type. Python will determine that based on the value you give.

```python
#Basic value assignment
x = 5
y = "hello"
z = 3.1


#Multiple assignment

a, b, c = 1, 2, 3

print("Type of x: ", type(x))
print("Type of y: ", type(y))
print("Type of z: ", type(z))



# variable reassignment

x = "Now I am a String"
print("Type of x: ", type(x))

```

## Input and Output
Python provides built-in functions for input/output operations. input() for input operations, print() output operations.

```python
#Basic Input
name = input("Enter your name: ")

#Basic Output
print("Hello", name)

#Formatted output
age = 25
height = 6
print(f"Name: {name}, Age: {age}, Height: {height}")
```

```python
#writing to a file

with open("myFile2.txt", "w") as file:
  file.write("This is a line of text")
  file.write("\nThis is another line of text")
  file.write(f"\nName: {name}, Age: {age}, Height: {height}")


#reading from a file

with open("myFile2.txt", "r") as f:
  content = f.read()
  print(content)
```

## Numbers
Python has several numerical types:int(integer), float(floating-point), complex. It also provides various operations and functions working with them.

```python
#integer operations

a = 5
b = 2

print("a + b = ", a + b)
print("a - b = ", a - b)
print("a * b = ", a * b)
print("a / b = ", a / b)

print("a ** b = ", a ** b) #Exponential
print("a % b = ", a % b) #Modulus



#float operations
x = 3.14
y = 2.71

print("x + y = ", x + y)
print(f"Round x to 1 decimal place: {round(x,1)}" )



#using math module

import math

print(f"Square root of 16: {math.sqrt(16)}")
print(f"Sin of 90 degrees: {math.sin(math.pi/2)}")
```
